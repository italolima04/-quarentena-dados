# Trabalhando com as Notas

```
#Notas das competências + Nota total da redação
notas_redacao = ['NU_NOTA_COMP1', 'NU_NOTA_COMP2', 'NU_NOTA_COMP3', 'NU_NOTA_COMP4', 'NU_NOTA_COMP5', 'NU_NOTA_REDACAO']
```

```
#Notas das 4 provas objetivas
notas_provas = ['NU_NOTA_CN', 'NU_NOTA_CH', 'NU_NOTA_MT', 'NU_NOTA_LC']
```

```
#Notas dos 5 tipos de avaliações do ENEM
notas_geral = ['NU_NOTA_CN', 'NU_NOTA_CH', 'NU_NOTA_MT', 'NU_NOTA_LC', 'NU_NOTA_REDACAO']
```

```
#Visualizando as maiores notas por competência.
maiores_notas = dados[notas_geral].max()
maiores_notas
```

```
#Tabela de dados com as notas p/ cada Participante
soma_notas = dados.groupby('NU_INSCRICAO').sum()[notas_geral]
soma_notas 
```

```
#As 10 melhores médias do Enem 2018
media_geral = (dados['NU_NOTA_CN'] + dados['NU_NOTA_CH'] + dados['NU_NOTA_MT'] + dados['NU_NOTA_LC'] + dados['NU_NOTA_REDACAO'])/5
#top_medias = dados.groupby('NU_INSCRICAO')[media_geral]
media_geral.sort_values(ascending=False).head(10)
```

```
#Eliminando dados que tiveram pelo menos uma nota auxente
dados_notas = dados[nota_geral].dropna()
dados_notas.columns = ['Natureza', 'Humanas', 'Matematica', 'Linguagens', 'Redacao']
dados_notas.head()
```

```
dados_notas.corr()
```
